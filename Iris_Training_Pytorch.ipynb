```python
from sklearn.datasets import load_iris
```

```python
import numpy as np
import pandas as pd
from torch.utils.data import DataLoader
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
dt = load_iris()
```

```python
dt.keys()
```

```python
X, Y = dt.data, dt.target
```

```python
dt.target_names
```

```python
model = nn.Sequential(
    nn.Linear(4, 30),
    nn.ReLU(),
    nn.Linear(30, 10),
    nn.ReLU(),
    nn.Linear(10, 3),
    # nn.Softmax(dim=1)
)
```

```python
X = torch.FloatTensor(X)
Y = torch.LongTensor(Y)
```

```python
model(X)
```

```python
torch.argmax(model(X), dim=1)
```

```python
lossfn = nn.CrossEntropyLoss()
optSgd = optim.SGD(model.parameters(), lr=0.01)
```

```python

for _ in range(200):
    optSgd.zero_grad()
    Yp = model(X)
    loss = lossfn(Yp, Y)
    loss.backward()
    optSgd.step()

    print(loss.item())
```

```python
for _ in range(100):
    optSgd.zero_grad()
    Yp = model(X)
    loss = lossfn(Yp, Y)
    loss.backward()
    optSgd.step()

    print(loss.item())
```
